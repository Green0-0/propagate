```python
import re
import matplotlib.pyplot as plt

# --- Paste your log data inside the triple quotes ---
log_data = """
"""
# ----------------------------------------------------

def moving_average(data, window_size):
    """
    Calculates a simple moving average for a list of numbers.
    For the first few points where a full window isn't available,
    it averages the points available so far.
    """
    if window_size <= 0:
        return data
    
    averages = []
    for i in range(len(data)):
        # Get the window of data to average
        # This window grows up to the window_size
        window = data[max(0, i - window_size + 1) : i + 1]
        
        # Calculate and append the average
        averages.append(sum(window) / len(window))
    return averages

# --- You can adjust this window size ---
# A larger window gives a smoother graph, but lags more.
# A smaller window is more responsive, but less smooth.
SMOOTHING_WINDOW_SIZE = 5
# ----------------------------------------


# Initialize lists to store the extracted data
iterations = []
average_rewards = []

# Define the regular expression to find iteration number and average reward
# This pattern looks for "IT <number>: Average <float>"
pattern = re.compile(r"IT (\d+): Average (\d+\.\d+)")

# Find all matches in the log data
matches = pattern.finditer(log_data)

# Extract data from each match
for match in matches:
    try:
        # Convert the captured groups to the correct type (int and float)
        iteration_num = int(match.group(1))
        avg_reward = float(match.group(2))
        
        #Append to our lists
        iterations.append(iteration_num)
        average_rewards.append(avg_reward)
    except ValueError as e:
        print(f"Could not parse line: {match.group(0)}. Error: {e}")

# Check if we actually found any data
if iterations and average_rewards:
    print(f"Successfully extracted {len(iterations)} data points.")
    
    # Calculate the smoothed rewards
    smoothed_rewards = moving_average(average_rewards, SMOOTHING_WINDOW_SIZE)
    
    # Create the plot
    plt.figure(figsize=(12, 7))
    
    # Plot the original data with low opacity
    plt.plot(iterations, average_rewards, marker='o', linestyle='-', markersize=3, alpha=0.3, label="Original Average Reward")
    
    # Plot the smoothed data
    plt.plot(iterations, smoothed_rewards, marker='', linestyle='-', linewidth=2, color='red', label=f"Smoothed Reward (Window={SMOOTHING_WINDOW_SIZE})")
    
    # Set titles and labels
    plt.title("Average Reward per Iteration (Smoothed)")
    plt.xlabel("Iteration")
    plt.ylabel("Average Reward")
    
    # Add a legend
    plt.legend()
    
    # Add a grid for better readability
    plt.grid(True)
    
    # Adjust layout to prevent labels from being cut off
    plt.tight_layout()
    
    # Display the plot
    print("Showing plot...")
    plt.show()
else:
    print("No data was extracted.")
    print("Please make sure to paste your log data into the 'log_data' variable in the script.")
```
